```python
# STEP 1: Convert data types
# convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# convert 'release_year' to numeric format
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
#STEP 2: Extract month and year from the 'date_added' column for time-based analysis.
#extract month and year from 'date_added'
df['date_added_month'] = df['date_added'].dt.month
df['date_added_year'] = df['date_added'].dt.year
```

```python
duration_null_df = df[df['duration'].isnull()]
duration_null_df
```

```python
# Identify rows with incorrect ratings that should be in the duration column
incorrect_ratings = ['74 min', '84 min', '66 min']
mask = df['rating'].isin(incorrect_ratings) & df['duration'].isnull()

# Move incorrect ratings to the duration column
df.loc[mask, 'duration'] = df['rating']

# Replace these values in the rating column with 'Not Rated' or another appropriate placeholder
df.loc[mask, 'rating'] = 'Not Rated'
```

```python
# STEP 3: Handle Missing Values
# Filling missing values in 'director' column with 'Unknown'
df['director'] = df['director'].fillna('Unknown', inplace=False)

# Filling missing values in 'cast' column with 'Unknown'
df['cast'] = df['cast'].fillna('Unknown', inplace=False)

# Filling missing values in 'cast' column with 'Unknown'
df['country'] = df['country'].fillna('Unknown', inplace=False)

# Filling missing values in 'date_added' column with the mode
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0], inplace=False)

# Filling missing values in 'date_added_month' column with the mode
df['date_added_month'] = df['date_added_month'].fillna(df['date_added_month'].mode()[0], inplace=False)

# Filling missing values in 'date_added_year' column with the mode
df['date_added_year'] = df['date_added_year'].fillna(df['date_added_year'].mode()[0], inplace=False)

# Filling missing values in 'ratings' column with 'Not rated'
df['rating'] = df['rating'].fillna('Not rated', inplace=False)
```

```python
# remove duplicates
df = df.drop_duplicates()
```

```python
# STEP 4: Clean Categorical Data
# standardise capitalization in categorical columns
df['type'] = df['type'].str.capitalize()
df['rating'] = df['rating'].str.upper()
```

```python
# STEP 5: Remove duplicates
df = df.drop_duplicates()
```

```python
#STEP 6: removing outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_out = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_out

df = remove_outliers(df, 'release_year')
```

```python
# remove any extra spaces from string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Creating new columns for duration to seperate movies

df['movie_duration_minutes'] = None

df.loc[df['type'] == 'Movie', 'movie_duration'] = df['duration']
df['movie_duration_minutes'] = df['movie_duration'].str.extract('(\d+)').astype(float)

df.drop(columns=['movie_duration'], inplace=True)

df.head()
```

```python
# Creating new columns for duration to seperate tv shows

df['tv_duration'] = None

df.loc[df['type'] == 'Tv show', 'tv_duration'] = df['duration']
df['tv_duration_seasons'] = df['tv_duration'].str.extract('(\d+)').astype(float)

df.drop(columns=['tv_duration'], inplace=True)

df.head()
```

```python
# 2. Use Clustering Techniques to Identify Different Groups of Users or Movies:

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Preparing data for clustering for release_year and movie_duration_minutes
df_clustering = df[['release_year', 'movie_duration_minutes']]
df_clustering = df_clustering.dropna()

# Standardize the features
scaler = StandardScaler()
df_clustering_scaled = scaler.fit_transform(df_clustering)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_clustering['cluster'] = kmeans.fit_predict(df_clustering_scaled)

#Reduce dimensions for visualisations
pca = PCA(n_components=2)
df_clustering_pca = pca.fit_transform(df_clustering_scaled)

df_clustering['pca1'] = df_clustering_pca[:,0]
df_clustering['pca2'] = df_clustering_pca[:,1]

#Plot the clusters
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_clustering, x='pca1', y='pca2', hue='cluster', palette='viridis')
plt.title('K-Means Clustering of Movies')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Findings
# Cluster 0: Older movies that have shorter duration
# Cluster 1: Recent movies that have average duration
# Cluster 2: Newest Movies that have longer duration
# Shows us that newer movies are much longer now and netflix may consider releasing more new longer movies.
```

```python
# 3. Test your hypotheses to verify any observations or conclusions.
# Hypthosis 1: Test whether movies and TV Shows have been realeased at different times and what trend in content production of the years

from scipy.stats import ttest_ind

release_year_movie = df[df['type'] == 'Movie']['release_year'].dropna()
release_year_tv_show = df[df['type'] == 'Tv show']['release_year'].dropna()

t_stat_release, p_val_release = ttest_ind(release_year_movie, release_year_tv_show)

print(f"T-statistic for release year: {t_stat_release}")
print(f"P-value for release year: {p_val_release}")


# Interpretation
if p_val_release < 0.05:
    print("Reject the null hypothesis for release year: There is a significant difference in the release years of movies and TV shows.")
else:
    print("Fail to reject the null hypothesis for release year: There is no significant difference in the release years of movies and TV shows.")
```

```python
# Hypthosis 2: Test the duration of movies and tv shows to under content consumption patterns.

duration_movie = df[df['type'] == 'Movie']['movie_duration_minutes'].dropna()
duration_tv_show = df[df['type'] == 'Tv show']['tv_duration_seasons'].dropna()

t_stat_duration, p_val_duration = ttest_ind(duration_movie, duration_tv_show)

print(f"T-statistic for duration: {t_stat_duration}")
print(f"P-value for duration: {p_val_duration}")

if p_val_duration < 0.05:
    print("Reject the null hypothesis for duration: There is a significant difference in the duration of movies and the number of seasons for TV shows.")
else:
    print("Fail to reject the null hypothesis for duration: There is no significant difference in the duration of movies and the number of seasons for TV shows.")
```
